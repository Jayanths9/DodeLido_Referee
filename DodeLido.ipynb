```python
# Importing Library Files
import albumentations as A
import cv2
from matplotlib import pyplot as plt
```

```python
# Creating a class to perform transformation
transform = A.Compose([
    #A.RandomCrop(width=256, height=256),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=1),
    A.RandomBrightnessContrast(p=0.2),
])
```

```python
# Reading image from the Google Collab Storage
image = cv2.imread("/content/dataset/0.png")
```

```python
# Converting image from Blue, Green, Red to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Perform transformation on the image
transformed = transform(image=image)
transformed_image = transformed["image"]
```

```python
# Display the Original and Transformed Image
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(1, 2, 2)
plt.imshow(transformed_image)
plt.axis('off')
plt.title("Transformed Image")
```
